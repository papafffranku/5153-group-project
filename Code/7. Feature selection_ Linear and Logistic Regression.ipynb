Set a threshold of at least 0.05 correlation to find significant numerical features:

```python
import numpy as np
import pandas as pd

# Filter the DataFrame to keep only columns with numeric data types
numeric_df = df.select_dtypes(include=[np.number])

# Define target columns that you want to display
targets = ['Anxiety', 'Depression', 'OCD', 'Insomnia']

# Calculate the correlation matrix for all numeric columns
corr_matrix = numeric_df.corr()

# Create a significant correlation threshold
significant_threshold = 0.05

# Iterate over each target, filtering and displaying significant correlations
for target in targets:
    if target in corr_matrix.columns:
        # Filter for significant correlations
        significant_corrs = corr_matrix[target][corr_matrix[target].abs() > significant_threshold]
        # Remove self-correlation by dropping the target itself from the results
        significant_corrs = significant_corrs.drop(target, errors='ignore')
        # Check if there are significant correlations to display
        if not significant_corrs.empty:
            print(f"Significant correlations with {target}:")
            display(significant_corrs.sort_values(ascending=False).to_frame())
        else:
            print(f"No significant correlations found for {target}.")
```

## 1.2 Categorical features: ANOVA Test

Find the categorical features.

```python
original_columns = set(df.columns)

numeric_columns = set(numeric_df.columns)

non_numeric_columns = original_columns - numeric_columns
non_numeric_columns_list = list(non_numeric_columns)

print(non_numeric_columns_list)
```

```python
# categorical_columns = ['While working', 'Music effects', 'Composer', 'Primary streaming service', 'Exploratory', 'Instrumentalist', 'Foreign languages', 'Fav genre'] # we will drop music effects as this also needs to be predicted
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Numerical targets
numerical_targets = ['Anxiety', 'Depression', 'OCD', 'Insomnia']

categorical_features = ['While working', 'Composer', 'Primary streaming service', 'Exploratory', 'Instrumentalist', 'Foreign languages', 'Fav genre']

results_anova = pd.DataFrame(columns=['Feature', 'Target', 'F-Value', 'P-Value'])

for target in numerical_targets:
    for feature in categorical_features:
        # Formulate the model
        model = ols(f'Q("{target}") ~ C(Q("{feature}"))', data=df).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)
        f_value = anova_table['F'][0]
        p_value = anova_table['PR(>F)'][0]
        # Store results in DataFrame
        results_anova = pd.concat([results_anova, pd.DataFrame({
            'Feature': [feature], 'Target': [target], 'F-Value': [f_value], 'P-Value': [p_value]
        })], ignore_index=True)

# Group the results by 'Target' and display each group as a separate DataFrame
grouped_results = results_anova.groupby('Target')

# Display results for each target, but only if they are significant (p-value < 0.05)
for target, group in grouped_results:
    significant_group = group[group['P-Value'] < 0.05]  # Filter for significance within each group
    if not significant_group.empty:
        print(f"Significant Results for {target}:")
        display(significant_group.sort_values(by='P-Value', ascending=True))
    else:
        print(f"No significant results for {target}.")
```
