## 1.1 frequency Mapping

```python
kg['Frequency [Pop]'].value_counts()
```

```python
import pandas as pd

# Define the frequency mapping
frequency_mapping = {
    "Never": 0,
    "Rarely": 1,
    "Sometimes": 2,
    "Very frequently": 3
}

# Identify the genre frequency columns
genre_frequency_columns = [col for col in kg.columns if 'Frequency [' in col]

# Apply the mapping to each frequency column
for col in genre_frequency_columns:
    kg[col] = kg[col].map(frequency_mapping)

# Create new columns for normalized frequencies
normalized_columns = {col: col + '_nor' for col in genre_frequency_columns}
for original_col, new_col in normalized_columns.items():
    kg[new_col] = kg[original_col]

# Normalize the new frequency values so that each row sums to 1 across the original frequency columns
kg[list(normalized_columns.values())] = kg[list(normalized_columns.values())].div(kg[genre_frequency_columns].sum(axis=1), axis=0)

# Handle any NaN values that may arise from dividing by zero (e.g., where all genre frequencies are zero)
kg[list(normalized_columns.values())] = kg[list(normalized_columns.values())].fillna(0)

# Display the first few rows to check the results
print(kg[[col for pair in zip(genre_frequency_columns, list(normalized_columns.values())) for col in pair]].head())
```

```python
# Metrics without missing values
metrics_no_missing = ['danceability', 'energy', 'loudness', 'mode', 'speechiness',
                      'acousticness', 'instrumentalness', 'liveness', 'valence',
                      'tempo', 'duration_ms', 'time_signature']

# Metrics with potential missing values
metrics_with_missing = ['metric_depression', 'metric_anxiety', 'metric_ocd', 'metric_insomnia']
```

```python
# Calculate the mean, skipping NA values only for specific columns
genre_averages = ly.groupby('genre').agg({**{metric: 'mean' for metric in metrics_no_missing},
                                          **{metric: lambda x: x.mean(skipna=True) for metric in metrics_with_missing}})
```

```python
# Reset the index to make 'genre' a column again
genre_averages.reset_index(inplace=True)
```

## 1.3 Weighted averages

```python
import pandas as pd

# Define the list of metrics to include in weighted averages
metrics_to_weight = ['danceability', 'energy', 'loudness', 'mode', 'speechiness',
                     'acousticness', 'instrumentalness', 'liveness', 'valence',
                     'metric_depression', 'metric_anxiety', 'metric_ocd', 'metric_insomnia']

# Normalize columns reference
normalized_columns = {col: col + '_nor' for col in genre_frequency_columns}

# Initialize a list to collect all the DataFrames
weighted_metrics_dfs = []

# Iterate through each normalized genre column in user data to calculate weighted metrics
for original_col, norm_col in normalized_columns.items():
    genre_name = original_col.split('[')[-1].split(']')[0].strip()
    if genre_name in genre_averages['genre'].values:
        # Create a temporary DataFrame to hold weighted metrics for the current genre
        temp_df = kg[norm_col].to_frame()
        for metric in metrics_to_weight:
            weighted_metric_name = f'{genre_name}_{metric}_weighted'
            # Calculate weighted metric and store in temp_df
            temp_df[weighted_metric_name] = temp_df[norm_col] * genre_averages.set_index('genre').at[genre_name, metric]

        # Append the temp_df without the original normalized frequency column to the list
        temp_df = temp_df.drop(columns=[norm_col])
        weighted_metrics_dfs.append(temp_df)

# Concatenate all weighted metric DataFrames horizontally
weighted_sums = pd.concat(weighted_metrics_dfs, axis=1)

# Sum the weighted metrics across all genres to get a total weighted value for each metric per user
for metric in metrics_to_weight:
    columns_to_sum = [col for col in weighted_sums.columns if metric in col]
    kg[f'weighted_{metric}'] = weighted_sums[columns_to_sum].sum(axis=1)
```
