```python
import pandas as pd

# Define the DataFrame structure
needed_columns = [
    'Country', 'Timestamp', 'Age', 'Primary streaming service', 'Hours per day',
    'While working', 'Instrumentalist', 'Composer', 'Fav genre', 'Exploratory',
    'Foreign languages', 'BPM', 'Frequency [Classical]', 'Frequency [Country]',
    'Frequency [EDM]', 'Frequency [Folk]', 'Frequency [Gospel]', 'Frequency [Hip hop]',
    'Frequency [Jazz]', 'Frequency [K pop]', 'Frequency [Latin]', 'Frequency [Lofi]',
    'Frequency [Metal]', 'Frequency [Pop]', 'Frequency [R&B]', 'Frequency [Rap]',
    'Frequency [Rock]', 'Frequency [Video game music]', 'Anxiety', 'Depression',
    'Insomnia', 'OCD', 'Music effects', 'Permissions'
]

countries = pd.DataFrame(columns=needed_columns)

# Data from data exploration
sg_genre_counts = {
    'Pop': 39,
    'K pop': 5,
    'R&B': 5,
    'Rock': 1
}

ind_genre_counts = {
    'Pop': 47,
    'K pop': 2,
    'R&B': 1
}

# Calculate total counts
total_sg = sum(sg_genre_counts.values())
total_ind = sum(ind_genre_counts.values())

# Create a dictionary for the row for Singapore and Indonesia
sg_row = {'Country': 'Singapore'}
ind_row = {'Country': 'Indonesia'}

# Set up default values for the other fields if necessary
default_values = {col: None for col in needed_columns if col not in ['Country']}
sg_row.update(default_values)
ind_row.update(default_values)

# Update the row dictionaries with the genre frequencies
for genre, count in sg_genre_counts.items():
    sg_row[f'Frequency [{genre}]'] = (count / total_sg) * 100

for genre, count in ind_genre_counts.items():
    ind_row[f'Frequency [{genre}]'] = (count / total_ind) * 100

# Append rows to the DataFrame
new_rows = pd.DataFrame([sg_row, ind_row])
countries = pd.concat([countries, new_rows], ignore_index=True)

# Display the new DataFrame
display(countries)
```

```python
import pandas as pd

# Identify the genre frequency columns
genre_frequency_columns = [col for col in countries.columns if 'Frequency [' in col]
countries[genre_frequency_columns] = countries[genre_frequency_columns].fillna(0)

# Create new columns for normalized frequencies
normalized_columns = {col: col + '_nor' for col in genre_frequency_columns}
for original_col, new_col in normalized_columns.items():
    countries[new_col] = countries[original_col]

# Normalize the new frequency values so that each row sums to 1 across the original frequency columns
countries[list(normalized_columns.values())] = countries[list(normalized_columns.values())].div(countries[genre_frequency_columns].sum(axis=1), axis=0)

# Handle any NaN values that may arise from dividing by zero (e.g., where all genre frequencies are zero)
countries[list(normalized_columns.values())] = countries[list(normalized_columns.values())].fillna(0)

# Display the first few rows to check the results
print(countries[[col for pair in zip(genre_frequency_columns, list(normalized_columns.values())) for col in pair]].head())
```

## 4 calculate Weighted averages

```python
countries_sg = countries[countries['Country']=='Singapore']
countries_ind = countries[countries['Country']=='Indonesia']
```

```python
### SINGAPORE ###
import pandas as pd

metrics_to_weight = ['danceability', 'energy', 'loudness', 'mode', 'speechiness',
                     'acousticness', 'instrumentalness', 'liveness', 'valence',
                     'metric_depression', 'metric_anxiety', 'metric_ocd', 'metric_insomnia']

normalized_columns = {col: col + '_nor' for col in genre_frequency_columns}

weighted_metrics_dfs = []

for original_col, norm_col in normalized_columns.items():
    genre_name = original_col.split('[')[-1].split(']')[0].strip()
    if genre_name in genre_averages_sg['genre'].values:
        temp_df = countries_sg[norm_col].to_frame()
        for metric in metrics_to_weight:
            weighted_metric_name = f'{genre_name}_{metric}_weighted'
            temp_df[weighted_metric_name] = temp_df[norm_col] * genre_averages_sg.set_index('genre').at[genre_name, metric]

        temp_df = temp_df.drop(columns=[norm_col])
        weighted_metrics_dfs.append(temp_df)

weighted_sums = pd.concat(weighted_metrics_dfs, axis=1)

for metric in metrics_to_weight:
    columns_to_sum = [col for col in weighted_sums.columns if metric in col]
    countries_sg[f'weighted_{metric}'] = weighted_sums[columns_to_sum].sum(axis=1)

display(countries_sg.head())
```
