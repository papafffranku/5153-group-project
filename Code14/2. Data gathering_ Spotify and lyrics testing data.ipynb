```python
cid='46fdb0714b7f41199db1c306d79168e4'
secret='cc44dadc45cc4c4a9b29d617db0c1602'
client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
sp = spotipy.Spotify(client_credentials_manager = client_credentials_manager)
```

```python
playlists = [
    ('Singapore', 'https://open.spotify.com/playlist/37i9dQZEVXbN66FupT0MuX?si=d549aa521d0e4970'),
    ('Indonesia', 'https://open.spotify.com/playlist/37i9dQZEVXbObFQZ3JLcXt?si=c468349990754ded')
]
```

```python
def get_playlist_tracks_info(playlist_url, country):
    results = sp.playlist_tracks(playlist_url)
    tracks = results['items']
    
    track_info = []
    for track in tracks:
        if track['track'] is not None:
            track_id = track['track']['id']
            artist_name = track['track']['artists'][0]['name']
            song_name = track['track']['name']
            
            # Get audio features
            audio_features = sp.audio_features([track_id])
            if audio_features:
                features = audio_features[0]
                # Append desired info to list
                track_info.append({
                    'Country': country,
                    'Artist': artist_name,
                    'Song': song_name,
                    'Danceability': features['danceability'],
                    'Energy': features['energy'],
                    'Key': features['key'],
                    'Loudness': features['loudness'],
                    'Mode': features['mode'],
                    'Speechiness': features['speechiness'],
                    'Acousticness': features['acousticness'],
                    'Instrumentalness': features['instrumentalness'],
                    'Liveness': features['liveness'],
                    'Valence': features['valence'],
                    'Tempo': features['tempo'],
                    'Duration_ms': features['duration_ms']
                })
    
    return track_info

# Loop through playlists and retrieve track information
all_tracks_info = []
for country, playlist_url in playlists:
    tracks_info = get_playlist_tracks_info(playlist_url, country)
    all_tracks_info.extend(tracks_info)

# Convert list of dictionaries to pandas DataFrame
df = pd.DataFrame(all_tracks_info)
```

```python
API = azapi.AZlyrics('google', accuracy=0.5)
```

```python
def get_lyrics(artist, title):
    try:
        API.artist = artist
        API.title = title
        lyrics = API.getLyrics(save=False)
        return lyrics
    except Exception as e:
        print(f"Error fetching lyrics for {title} by {artist}: {e}")
        return None
```

```python
df['lyrics'] = df.apply(lambda row: get_lyrics(row['Artist'], row['Song']), axis=1)
```

```python
genre_mapping = {
    'Taylor Swift': 'Pop',
    'Ariana Grande': 'Pop',
    'Sabrina Carpenter': 'Pop/R&B',
    'Jung Kook': 'K-pop/Rap',
    'Hozier': 'Rock',
    'BABYMONSTER': 'Pop',
    'Tate McRae': 'Pop',
    'LE SSERAFIM': 'Pop',
    'SZA': 'R&B',
    'Jimin': 'R&B',
    'Artemas': 'Pop',
    'Henry Moodie': 'Pop',
    'Djo': 'Pop'
}

def map_genre(row):
    artist = row['Artist']
    if artist in genre_mapping:
        return genre_mapping[artist]
    else:
        return 'Unknown'  # Handle cases where genre is not defined

# Apply mapping to create new 'Genre' column
df['Genre'] = df.apply(map_genre, axis=1)
```
